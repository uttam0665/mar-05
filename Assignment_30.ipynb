```python
"""Q1. How can you create a Bokeh plot using Python code?"""
#Ans:-

"""Bokeh is a Python library that provides elegant and interactive data visualization capabilities for modern web browsers. Here's an example of how to create a simple Bokeh plot using Python code:"""

#1.Import the necessary libraries:
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
#2.Create a Bokeh figure object:
p = figure(title="My Plot", x_axis_label="X Axis", y_axis_label="Y Axis")
#Add data to the figure:
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.line(x, y)
#Display the plot:
output_notebook()
show(p)
from bokeh.io import output_file, show
output_file("plot.html")
show(p)
```

```python
"""Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example."""
#Ans:-
"""In Bokeh, glyphs are visual markers that represent data points on a plot. Glyphs can be added to a Bokeh plot using the glyph() method of a figure object. There are many different types of glyphs available in Bokeh, including circles, squares, lines, and more."""
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
p = figure(title="My Plot", x_axis_label="X Axis", y_axis_label="Y Axis")
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.circle(x, y, size=10, color="navy", alpha=0.5)
output_notebook()
show(p)
```

```python
"""Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?"""
#ANS:-
"""To customize the appearance of a Bokeh plot, you can use the properties of the various Bokeh objects that make up the plot, such as the axes, title, and legend. Here are some examples:
1.Changing the font size and color of the title:"""

from bokeh.plotting import figure, output_file, show
p = figure(title="My Plot", plot_width=400, plot_height=400)
p.title.text_color = "blue"
p.title.text_font_size = "20pt"
p.xaxis.axis_line_color = "red"
p.xaxis.axis_line_width = 3
p.yaxis.axis_line_color = "green"
p.yaxis.axis_line_width = 2

p.xaxis.axis_label = "X-Axis Label"
p.xaxis.axis_label_text_color = "purple"
p.xaxis.axis_label_text_font_size = "16pt"
p.yaxis.axis_label = "Y-Axis Label"
p.yaxis.axis_label_text_color = "orange"
p.yaxis.axis_label_text_font_size = "14pt"

p.scatter(x=[1,2,3], y=[1,2,3], legend_label="Scatter Plot")
p.legend.location = "top_left"
p.legend.label_text_font_size = "14pt"

```
